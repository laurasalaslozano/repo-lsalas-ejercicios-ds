Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
plt.style.use("seaborn-v0_8-white")

fig, ax = plt.subplots(2, 2, figsize = (15,15))
```

1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

```python
top = df_cities.sort_values("population_total", ascending = False).head()

x = top["city"]
y = top["population_total"]

ax1 = ax[0,0]
ax1.bar(x, y, color = ["blue", "grey", "olive", "brown"], alpha = 0.5)

ax1.set_title("Ciudades ms pobladas de California")
ax1.set_xlabel("Ciudad")
ax1.set_ylabel("Poblacin")


for indice, valor in enumerate(y):
    ax1.text(indice, valor, str(valor), ha = "center")

```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")

```python
x1 = df_cities["area_total_km2"]

ax2 = ax[0,1]

ax2.set_xlabel("Superficie Km2", fontname = "Bradley hand ITC", fontsize = 18)
ax2.set_ylabel("Ciudad", fontname = "Bradley hand ITC", fontsize = 18)
ax2.legend(["Hist Area"], frameon = True)

ax2.hist(x1,
         bins = 50,
         color = "steelblue");
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
x2 = df_cities["area_water_km2"]
y2 =df_cities["population_total"]

ax3 = ax[1,0]

ax3.set_title("Poblacin respecto al rea de agua")
ax3.set_xlabel("rea de agua - Km2")
ax3.set_ylabel("Poblacin")

props_flecha = {
    "arrowstyle": "->",
    "connectionstyle": "angle3",
    "color": "k"
}

ax3.scatter(x2, y2, alpha = 0.3);
```

4. El boxplot de la altura de las ciudades.

```python
x3 = df_cities["elevation_m"]

ax4 = ax[1,1]

ax4.boxplot(x3, whis = 2, vert = False);
```
