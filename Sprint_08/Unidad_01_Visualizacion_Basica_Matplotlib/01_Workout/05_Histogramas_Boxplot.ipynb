# Visualizacin con Matplotlib: Histogramas y diagramas de caja

Vamos a ver de forma sencilla como hacer los grficos que nos generaban las funciones que empleamos en las visualizaciones analiticas del sprint anteior. En concreto, veremos los histogramas y los boxplots:

````python
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)
````

### Histogramas

Tan sencillo como:

````python
plt.hist(data);
````

````python
fig = plt.figure(figsize = (5,4))
ax = plt.axes()
ax.hist(data);
````


La funcin ``hist()`` tiene muchas opciones para ajustar tanto el clculo como la visualizacin; aqu tienes un ejemplo de un histograma ms personalizado (en el que tocamos el nmero de columnas `bins`, la transparencia y el color):


````python
plt.hist(data,
         bins = 50,
         alpha = 1,
         color = "steelblue"
         );
````

La transparencia es particularmente interesante para comparar histogramas o mostrar dicha comparacin (y aprovecho para ensearte un truco para cuando tengas argumentos repetidos que poner en tus llamadas a funcin):

````python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(alpha = 0.3, bins = 20)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
````

Es tambin el momento de ensearte a capturar los lmites de esos bines:

````python
fig = plt.figure()
ax = plt.axes()
limites = ax.hist(x1, alpha = 0.3, bins = 30)
````

Limites es una tupla en la que el primer valor son los conteos y el segundo los lmites de las barras:

````python
limites[0]
````

````python
limites[1]
````


### Boxplots

Los diagramas de caja eran esos diagramas en los que resumamos las medidas de posicin y aadamos una posible formar de obtener valores anmalos, extremos o outliers. En forma de diagrama:


```text
     Q1-1.5IQR   Q1  mediana  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier             <----------->            fliers/outlier
                       IQR

```

Donde IQR es el rango intercuartilico (la diferencia entre el valor de Q3 o tercer cuartil o percentil-75 y el valor de Q1 o primer cuartil, o percentil-25)

En matplotlib:

````python
plt.boxplot(data, whis = 1.5);
````

En el caso de matplotlib empleamos por defecto diagramas de caja verticales. Pero podemos cambiarlo:

````python
plt.boxplot(data, vert = False);
````

Y podemos mostrar varios juntos:

````python
plt.boxplot([x1,x2,x3]);
````
