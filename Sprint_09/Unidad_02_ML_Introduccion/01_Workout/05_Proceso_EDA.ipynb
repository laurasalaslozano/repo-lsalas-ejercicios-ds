```python
# Recuperemos lo hecho en la sesin anterior

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from bootcampviztools import pinta_distribucion_categoricas, plot_categorical_numerical_relationship, plot_grouped_histograms
from sklearn.model_selection import train_test_split


# Cdigo extra para los tamaos de las fuentes y los "ticks" de los ejes, esto no lo vimos al hablar de matplotlib as que una extra
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)


df_inmo = pd.read_csv("./data/ejemplo_housing.csv")

train_set, test_set = train_test_split(df_inmo, test_size=0.2, random_state=42)
```

### 1. Limpieza y transformacin

Deberamos acometer la limpieza y transformacin, pero como ya vimos no hay valores nulos y los datos parecen bastante limpios

```python
df_inmo.info()
```

Unicamente parece interesante convertir "mean_income" en una variable categrica, porque ya de por s se ha intentado hacer un escalado con la misma. Esto es algo que ya hicimos varias veces cuando analizamos datos, recurriendo a la funcin `cut` de pandas. Hagmoslo:


```python
train_set["income_cat"] = pd.cut(train_set["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
train_set["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
plt.show()
```

[NOTA: Importante, toda transformacin que se haga en las variables de train habr que hacerlas en los datos de test antes de poder aplicarles el modelo]

#### 2.2 Variables numricas y target

Por simplificar, directamente iremos a las correlaciones de las variables numricas con la variable target y pintando las ms significativas:

```python
corr_matrix = train_set.corr(numeric_only= True)
corr_matrix["median_house_value"].sort_values(ascending = False)
```

Como era de esperar "median_income" es claramente la que tiene ms correlacin, y podramos descartar poblacin y las geogrficas que hemos visto que hasta cierto punto estn contenidas en la categrica de cercana al mar (quizs la latitud podramos quedrnosla)

```python
columnas = corr_matrix["median_house_value"][corr_matrix["median_house_value"] > 0.07].index.to_list()
```

```python
sns.pairplot(train_set[columnas]);
```

Como era de esperar grficamente la "posible" correlacin ms potente es con "median_income". Pero en este caso no vamos a descartar las otras, y nos quedaremos con la categrica en vez de con "median_income". 

### 2.3 Crear otras variables

En este caso, podramos pensar que quizs algunas variables combinadas puedan ser interesantes por ejemplo:  
1. El nmero medio de habitaciones por casa ("total_rooms"/"households")  
2. El nmero de dormitorios en funcin del total de habitaciones ("total_bedrooms"/"total_rooms")
3. O el nmero medio de personas por casas ("population"/"housholds")

Creemos estas variables y veamos su correlacin con el target

```python
train_set["rooms_per_house"] = train_set["total_rooms"] / train_set["households"]
train_set["bedrooms_ratio"] = train_set["total_bedrooms"] / train_set["total_rooms"]
train_set["people_per_house"] = train_set["population"] / train_set["households"]
```

```python
corr_matrix = train_set.corr(numeric_only=True)
corr_matrix["median_house_value"].sort_values(ascending=False)
```

Hmm, interesante porque hemos encontrado una "mejor" correlacin en habitaciones por casa "rooms_per_house" pero ms interesante es la nueva variable ratio de dormitorios, que presenta una contraintuitiva "anticorrelacin". Es decir a menor ratio de dormitorios por habitaciones mayor precio.
